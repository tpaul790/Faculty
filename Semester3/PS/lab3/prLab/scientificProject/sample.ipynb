```python
from operator import truediv
from random import choices, sample, randint
from math import comb,perm

def generate()->list[str]:
    return 5*["r"]+3*["a"]+2*["v"]

def alege():
    return sample(generate(),3)

def simulare(l):
    k = 3
    return sample(l, k)

def verificaA(l):
    if "r" in l:
        return True
    return False

def verificaAB(l):
    if "r" in l:
        if len(set(l)) == 1:
            return True
    return False

def probabilityAB(times:int):
    cntA = 0
    cntAB = 0
    for _ in range(times):
        if verificaAB(alege()):
            cntAB += 1
        if verificaA(alege()):
            cntA += 1
    return cntAB/cntA
    

print(probabilityAB(5000))

```
