```python
from email.utils import decode_params

from scipy.stats import bernoulli, binom
from matplotlib.pyplot import bar, show, hist, grid, legend, xticks
#pr1 a)
def deplasare(pasi:int, p:float) -> int:
    poz = [0]
    for _ in range(pasi):
        x = bernoulli.rvs(p)
        poz.append(poz[-1]+verificarePas(x))
    return poz

def verificarePas(x:int):
    return 2*x-1

deplasare(pasi=5, p = 0.5)
```

```python
#b)
def simulare(times:int, pasi:int, p:float):
    data = []
    for _ in range(times):
        l = deplasare(pasi = pasi, p = p)
        data.append(l[-1])
    return data
def histograma(pasi:int, p:float):
    data = simulare(times = 1000 ,pasi=pasi, p = p)
    bin_edges = [k+0.5 for k in range(-pasi-1,pasi+1)]
    hist(data, bin_edges, density = True, rwidth = 0.9, color = 'green', edgecolor = 'black',alpha = 0.5, label = 'frecvente relative')
    
histograma(pasi = 7, p = 0.5)
```

```python
#c)
def simulare(times:int, pasi:int, p:float, nrPuncte:int):
    data = []
    for _ in range(times):
        l = deplasare(pasi = pasi, p = p)
        data.append(l[-1]%nrPuncte)
    return data

def histograma(pasi:int, p:float, nrPuncte:int):
    data = simulare(times = 1000 ,pasi=pasi, p = p, nrPuncte = nrPuncte)
    bin_edges = [k+0.5 for k in range(-1,nrPuncte)]
    hist(data, bin_edges, density = True, rwidth = 0.9, color = 'green', edgecolor = 'black',alpha = 0.5, label = 'frecvente relative')
    
histograma(pasi = 10, p = 0.5, nrPuncte = 6)
```
