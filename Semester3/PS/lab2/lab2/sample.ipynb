```python
import random

def generate(n:int):
    list = []
    for i in range(n):
        list.append(random.randint(1,365))
    return list
    
    
def simulate(pers:int, times:int, prob = False):
    if prob:
        p = 1
        for k in range(pers):
            p*=(365-k)/365
        return 1-p
    else:
        real = 0
        for k in range(1,times):
            l = generate(pers)
            cnt = 0
            for i in range(1, pers):
                for j in range(i+1, pers):
                    if l[i] == l[j]:
                        cnt += 1
            if cnt > 0:
                real += 1
        return real/times

print(simulate(23,1000))
```

```python
from matplotlib.pyplot import plot, grid, title, show
title("Probability")
xs = range(2,51)
ys = [simulate(k, 1000,True) for k in xs]
plot(xs, ys,'r*')
grid()
show()
```

```python
#2
#a)
from matplotlib.pyplot import axis, plot, show
from random import random
from math import dist
import math

axis('square')
axis((0,1,0,1))
def generate(number:int):
    CN = [0.5,0.5]
    A,B,C,D = [0,1],[1,0],[0,0],[1,1]
    cntR=0
    for _ in range(1,number):
        F = [random(),random()]
        if math.dist(F, CN)<min([math.dist(F,A),math.dist(F,B),math.dist(F,C),math.dist(F,D)]):
            plot(F[0],F[1],'ro')
            cntR += 1
        else:
            plot(F[0],F[1],'bo')
    return cntR/number
rez = generate(2000)
show()
print('Frecventa relativa: ',rez)
```
