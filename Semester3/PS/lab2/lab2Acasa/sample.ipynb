```python
import math
from random import randint


def generare(n:int) -> list[int]:
    l = []
    for i in range(0,n+1):
        l.append(randint(1,365))
    return l

def simulari(n:int, times:int):
    rez = 0
    for _ in range(times):
        l = generare(n)
        cnt = 0
        for i in range(n):
            for j in range(i+1,n):
                if l[i] == l[j]:
                    cnt += 1
                    break
        if cnt > 0:
            rez += 1
    return rez/times
simulari(23,1000)
```

```python
def probabilitate(n:int):
    p = 1
    for i in range(n):
        p *= (365-i)/365
    return 1-p

probabilitate(23)
```

```python
from matplotlib.pyplot import plot, grid, title, show
title("Plot test")
xs = range(2,51)
yx = [probabilitate(n) for n in xs]
plot(xs, yx, 'r*')
grid()
show()
```

```python
from matplotlib.pyplot import axis, plot, show
from random import random
from math import dist, pi
axis('square')
axis((0,1,0,1))
def generarePuncteInCercTangentLaturilorPatratului(n:int):
    centru = [0.5,0.5]
    raza = 0.5
    #distanta de la punct pana la centru trebe sa fie <= raza
    cntR = 0
    for i in range(n):
        punct = [random(), random()]
        if dist(punct, centru) < raza:
            plot(punct[0], punct[1], 'r*')
            cntR += 1
        else:
            plot(punct[0], punct[1], 'b*')
    return cntR/n

print("Probabilitatea relativa este:",generarePuncteInCercTangentLaturilorPatratului(5000))
print("Probabilitatea geometrica este:", pi * 0.5 ** 2/1)
show()
```

```python
from matplotlib.pyplot import axis, plot, show
from random import random
from math import dist, pi
axis('square')
axis((0,1,0,1))
def generarePuncteInCercMaiAproapeDeCentruDecatDeVarf(n:int):
    centru = [0.5,0.5]
    A,B,C,D = [1,0],[1,1],[0,1],[0,0]
    cntR = 0
    for i in range(n):
        punct = [random(), random()]
        if dist(punct, centru) < min([dist(punct,A),dist(punct,B),dist(punct,C),dist(punct,D)]):
            plot(punct[0], punct[1], 'r*')
            cntR += 1
        else:
            plot(punct[0], punct[1], 'b*')
    return cntR/n

print("Probabilitatea relativa este:",generarePuncteInCercMaiAproapeDeCentruDecatDeVarf(5000))
print("Probabilitatea geometrica este:", (1-4*(0.5**2)/2)/1)
show()
```
